```python
import numpy as np
import scipy.stats as stats
data=[12,15,14,10,18,20,22,25,14,17]
```

```python

range_value=max(data)-min(data)
print(f"Range:{range_value}")
```

```python

variance_value=np.var(data,ddof=0)
sample_variance_value=np.var(data,ddof=1)
print(f"Population variance:{variance_value}")
print(f"Sample varinace:{sample_variance_value}")
```

```python

std_dev_value=np.std(data,ddof=0)
sample_std_dev_value=np.std(data,ddof=1)
print(f"Population standard deviation:{std_dev_value}")
print(f"Sample standard deviation:{sample_std_dev_value}")
```

```python

q1=np.percentile(data,25)
q3=np.percentile(data,75)
iqr_value=q3-q1
print(f"interquartile range:{iqr_value}")
```

Frompyfunc

```python
import numpy as np
def myadd(x,y):
  return x+y
myadd=np.frompyfunc(myadd,2,1)
print(myadd([1,2,3,4],[5,6,7,8]))
```

## Rounding Decimals

### Truncation-trunc() #Remove the decimal

```python
import numpy as np
arr=np.trunc([-3.1666,3.6667])
print(arr)                      #trunc() & fix() act as same
```

### fix()

```python
import numpy as np
arr=np.fix([-3.1666,3.6667])
print(arr)
```

### Rounding-around()

```python
import numpy as np
arr=np.around(3.1666, 2)         #Rounding decimal points
print(arr)
```

### Floor()-

```python
import numpy as np
arr=np.floor([-3.1666,3.6667])
print(arr)                      #rounds off decimal to nearest lower integer
```

### Ceil

```python
import numpy as np
arr=np.ceil([-3.16667,3.6667])
print(arr)
```

## Finding Union

Find the intersection of th following two set arrays

```python
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([3,4,5,6])
x=np.intersect1d(arr1,arr2,assume_unique=True)
print(x)
```

### Finding Difference

```python
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([3,4,5,6])
newarr=np.setdiff1d(arr1,arr2,assume_unique=True)
print(newarr)
```

### Finding Sytmmetric Difference

```python
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([3,4,5,6])
x=np.setxor1d(arr1,arr2,assume_unique=True)
print(x)
```
