##Concat
It is used to concat two or more dataframes

```python
tcs={'id':[101,102,103,104],
     'name':['Midun','dipin','Jose','Rahul'],
     'age':[25,26,27,28]}
wipro={'id':[101,102,103,104],
       'name':['dev','sreenath','sreerag','shafeel'],
       'age':[24,24,27,26]}
tcs_emp=pd.DataFrame(tcs)
wipro_emp=pd.DataFrame(wipro)
pd.concat([tcs_emp,wipro_emp])
```

```python
#ignore index-To get common index after concatenation
pd.concat([tcs_emp,wipro_emp],ignore_index=True)
```

```python
df=pd.concat([tcs_emp,wipro_emp],keys=['tcs','wipro'])#keys-To provide seperate keys for each concatenate data frames
df
```

```python
#concatenating vertically
df=pd.concat([tcs_emp,wipro_emp],axis=1)
df
```

```python
#cocatinating a dataframe with a series
salary=pd.Series([65000,75000,15000,25000],name='salary')
pd.concat([tcs_emp,salary],axis=1)
```

## Merge

```python
tcs={'id':[101,102,103,104],
     'name':['Midun','dipin','Jose','Rahul'],
     'age':[25,26,27,28]}
wipro={'id':[101,102,103,104],
       'name':['dev','sreenath','sreerag','shafeel'],
       'age':[24,24,27,26]}
tcs_emp=pd.DataFrame(tcs)
wipro_emp=pd.DataFrame(wipro)

tcs_emp=pd.DataFrame(tcs)
tcs_emp


```

```python
wipro_emp=pd.DataFrame(wipro)
wipro_emp
```

```python
pd.merge(tcs_emp,wipro_emp,on='id')
```

```python
pd.merge(tcs_emp,wipro_emp,on='id',suffixes=['left','right'])
```

```python
pd.merge(tcs_emp,wipro_emp,on='id',how='outer')
```

```python
pd.merge(tcs_emp,wipro_emp,on='id',how='outer',indicator=True)
```

```python
pd.merge(tcs_emp,wipro_emp,on='id',how='right') #right join
```

```python
pd.merge(tcs_emp,wipro_emp,on='id',how='left') #left join
```

##Cross tab or contingency table
In pandas, pd.crosstab() is used to compute a cross-tabulation of two or more factors. It’s often used for analyzing categorical variables, similar to a contingency table in statistics.


```python
import pandas as pd
df=pd.read_csv("/content/survey_crosstab.csv")
df
```

```python
ct = pd.crosstab(df['Gender'], df['Preferred_Device'])
print(ct)
```

```python

# Cross-tab of AgeGroup vs Preferred_Device with totals:

ct2 = pd.crosstab(df['AgeGroup'], df['Preferred_Device'], margins=True, margins_name='Total')
print(ct2)
```
